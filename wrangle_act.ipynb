## Cleaning data ##

Now, I will do the cleaning process considering the problems cited in the topic above.

### 1st: Twitter Archive ###

First, let's make a copy of the original dataframe:

```python
twitter_archive_clean = twitter_archive.copy()
```

1. **Quality Issue**: We can see that the column source has a '< a href =' in the starg of the string and ' rel="nofollow">Twitter for iPhone< /a>' in the end of it, so it does not present the source in a straightforward way. Those parts of the string could be removed.

>**Define**: Let's cut the string parts that we do not want to remain. We need to remove 9 characters from the start of the string and 39 characters from the end of the string.

```python
#Code
twitter_archive_clean['source'] = [item[9:-39] for item in twitter_archive_clean['source']]

#Test
twitter_archive_clean['source'].sample(5)
```

Now we can see only the url in the source column, as shown above.

2. **Quality Issue**: We can see, using the info function, that the column expanded_urls has 59 null values (the data has 2.356 rows, and this column has only 2.297 non-null values), which is a quality problem.
>**Define**: In this case we don't have much to do since we don't know what are the missing urls. We could drop the missing value lines, but that would make us lose other important pieces of information of the datarame, inclus=ding more relevant ones than the urls.

3. **Quality Issue**: Timestamp and retweeted_status_timestamp data formats are string and should be datetime. 
4. **Quality Issue**: Tweet_id, in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id and retweeted_status_user_id formats are int and should be string.  
>**Define**: Let's change the data types using astype function.

```python
#Code
twitter_archive_clean['tweet_id'] = twitter_archive_clean['tweet_id'].astype('str')
twitter_archive_clean['in_reply_to_status_id'] = twitter_archive_clean['in_reply_to_status_id'].astype('int', errors = 'ignore').astype('str')
twitter_archive_clean['in_reply_to_user_id'] = twitter_archive_clean['in_reply_to_user_id'].astype('int', errors = 'ignore').astype('str')
twitter_archive_clean['retweeted_status_id'] = twitter_archive_clean['retweeted_status_id'].astype('int', errors = 'ignore').astype('str')
twitter_archive_clean['retweeted_status_user_id'] = twitter_archive_clean['retweeted_status_user_id'].astype('int', errors = 'ignore').astype('str')
twitter_archive_clean['timestamp'] = twitter_archive_clean['timestamp'].astype('datetime64')
twitter_archive_clean['retweeted_status_timestamp'] = twitter_archive_clean['retweeted_status_timestamp'].astype('datetime64')

#Test
print(twitter_archive_clean.info())
twitter_archive_clean.head()
```

Now we can see that the columns have the right data types using the table above and the function info.

5. **Quality Issue**: It appears that the maximum values in rating_numerator and rating_denominator were typed wrong.
>**Define**: Let's check for occurencies in which the denominator is not 10 and replace them for 10. In the case of the denominator, let's look for larger numbers than normal and replace them (in the case of th observed maximum numbet, I believe it was supposed to be 17.76.

```python
#Code for denominator
black_list_1 = twitter_archive_clean[twitter_archive_clean['rating_denominator'] != 10]

for i in black_list_1:
    twitter_archive_clean['rating_denominator'].replace(black_list_1[i], 10, inplace = True)

#Test for denominator
twitter_archive_clean['rating_denominator'].describe()    
```

```python
#Code for numerator
black_list_2 = twitter_archive_clean[twitter_archive['rating_numerator'] >= 100]['rating_numerator'].unique()

for i in black_list_2:
    if i<1000:
        twitter_archive_clean['rating_numerator'].replace(i, i/10, inplace = True)
    elif i >= 1000:
        twitter_archive_clean['rating_numerator'].replace(i, i/100, inplace = True)

#Test
twitter_archive_clean['rating_numerator'].describe()    
```

Using the describe functions we can see that in both columns the max value makes more sense now.

6. **Quality Issue**: Some tweets could be retweets. We can remove them from all the dataframes using the id's that appear in the retweeted_status_id column.
>**Define**: I will remove from all dataframes the items whose ids are present in the retweeted_status_id column. I will make a copy of the other two dataframes to do that. After this is done, we won't neet de retweet related columns anymore, so I will drop them.

```python
image_predict_clean = image_predict.copy()
tweet_counts_clean = tweet_counts.copy()
```

```python
#Code
retweets = [str(int(i)) for i in twitter_archive['retweeted_status_id'].dropna().unique()]

for i in retweets:
    twitter_archive_clean.drop(twitter_archive_clean[twitter_archive_clean['tweet_id'] == i].index, inplace = True)
    image_predict_clean.drop(image_predict_clean[image_predict_clean['tweet_id'].astype('str') == i].index, inplace = True)
    tweet_counts_clean.drop(tweet_counts_clean[tweet_counts_clean['tweet_id'].astype('str') == i].index, inplace = True)

twitter_archive_clean.drop(columns=['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], inplace = True)

#Test
print(twitter_archive_clean.shape[0], twitter_archive.shape[0])
print(image_predict_clean.shape[0], image_predict.shape[0])
print(tweet_counts_clean.shape[0], tweet_counts.shape[0])
twitter_archive_clean.head()
```

Using the function shape we can see that the twitter_archive_clean had 2,356 records, and now has 2,244. The image_predict_clean dataframe had 2,075 records and now has 1,974. The tweet_counts_clean had 2,328 records and now has 2,218. All this because we removed the retweets. We can also see in the table above that the columns we removed are not there anymore.

7. **Tidyness Issue**: Dog "stage" is distributed in four columns in the dataframe (doggo, floofer, pupper, puppo). This could be turned into only one column with the dog's stage information. In case of tweets with more than one dog stage, one of them will nedd to be dropped.
>**Define**: I am going to use the melt function to create only one column to dog's 'stage', and then I will remove one of the stages for cases with more than one stages reported.

```python
#Code

#First I will create the columns with the dog stage (name of columns of dog stages) and stage (actual stage od that tweet)
twitter_archive_clean = pd.melt(twitter_archive_clean, id_vars=['tweet_id','in_reply_to_status_id','in_reply_to_user_id','timestamp','source','text','expanded_urls','rating_numerator','rating_denominator','name'],
 var_name='dog_stage', value_name='stage')

#Now I will criate an auxiliar array that will receive a random integer number from 1 to 4 if the stage is different of None
# and 5 if the stage is equal to None. I'll add this array to the dataframe and sort the dataframe based on it, using ascending
#order. This way, when I drop the duplicates of ids I will specify to mantain the first occurence, which will be None only if
#there is no other stage reported to that tweet.

import random

aux = []
for i in twitter_archive_clean['stage']:
    if i == 'None':
        aux.append(5)
    else:
        aux.append(random.randint(1,5))

twitter_archive_clean['aux_stage'] = aux

#Sorting values as explained before
twitter_archive_clean.sort_values('aux_stage', ascending=True, inplace = True)

#Dropping duplicates
twitter_archive_clean.drop_duplicates(subset=['tweet_id'], keep='first', inplace = True)

#Removing comlumns that will no longer be used
twitter_archive_clean.drop(columns=['dog_stage', 'aux_stage'], inplace = True)

#Test
twitter_archive_clean.shape[0]  
```

As the final number of rows of the twitter_archive_clean is 2,244, we can see that we removed all duplicates from the melt process.

### 2nd: Image Predictions ###

1. **Quality Issue**: tweet_id format is int and should be string.
>**Describe**: I will change tweet_id type to string.

```python
#Code
image_predict_clean['tweet_id'] = image_predict_clean['tweet_id'].astype('str')

#Test
image_predict_clean['tweet_id'].describe()
```

The describe funtion shows that the tweet_id column is now a string.

2. **Quality Issue**: Some of the tweets are not of dogs, and we should remove them.
3. **Quality Issue**: Some of the tweets are retweets and we should remove them.
4. **Tidyness Issue**: This is not exactly a problem, but in order to really use the data as information we are going to have to choose the best model to each tweet based on the confidence interval to determine which is the dog's breed. This will result in two columns: one with the name of the chosen model and one with the dog's breed predicted by that model. After, we can drop the initial model related columns. 
>**Describe**: First, I will create a column that has the best prediction model in terms of confidence interval, and another column that has the dog's breed chosen by that model. Then, I will remove the rows in which p1_dog or p2_dog or p3_dog (depending on which is the best model) is False, and then I will drop those columns and the model related columns (except for the new ones created) because they won't be used anymore. The retweets were already removed in previous steps.

```python
#Code
#First I will create some auxiliar arrays to get the values of the best models
best_model = []
best_c_interval = []
best_dog_breed = []
best_is_dog = []

#By iterating the dataframe, I get the data of the best model of each row and store it into the auxiliar arrays
for row in image_predict_clean.iterrows():

    aux = max([row[1]['p1_conf'], row[1]['p2_conf'], row[1]['p3_conf']])
    
    if row[1]['p1_conf'] == aux:
        best_model.append('p1')
        best_c_interval.append(row[1]['p1_conf'])
        best_dog_breed.append(row[1]['p1'])
        best_is_dog.append(row[1]['p1_dog'])
    elif row[1]['p2_conf'] == aux:
        best_model.append('p2')
        best_c_interval.append(row[1]['p2_conf'])
        best_dog_breed.append(row[1]['p2'])
        best_is_dog.append(row[1]['p2_dog'])
    else:
        best_model.append('p3')
        best_c_interval.append(row[1]['p3_conf'])
        best_dog_breed.append(row[1]['p3'])
        best_is_dog.append(row[1]['p3_dog'])

#Here, I add the auxiliar arrays to the dataframe
image_predict_clean['best_model'] = best_model
image_predict_clean['dog_breed'] = best_dog_breed
image_predict_clean['is_dog'] = best_is_dog

#Here I drop the rows of tweets that do not refer to dogs
image_predict_clean = image_predict_clean[image_predict_clean['is_dog'] == True]

#Here I drop the columns that we will not use anymore
image_predict_clean.drop(columns=['p1_conf', 'p2_conf', 'p3_conf', 'p1', 'p2', 'p3', 'p1_dog', 'p2_dog', 'p3_dog', 'is_dog'], inplace = True)

#Test
print(image_predict_clean.shape[0])
image_predict_clean.head()
    
```

We can see in the table above the final format of our image_predict dataframe after cleaning, exaclty as expected.
